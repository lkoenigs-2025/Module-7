# Part 2 — Build an Abstract Parent Class

Create an abstract class called `Animal`.

Requirements:

1. `Animal` must inherit from `ABC`.
2. `Animal` must have a class attribute named `total_animals`, starting at `0`.
3. The constructor must accept `name` and `age`.
4. The constructor must save those as instance attributes.
5. The constructor must increase `total_animals` by 1.
6. Add a concrete method named `show_info()` that returns a string such as `Buddy is 4 years old.`
7. Add a `@classmethod` named `get_total_animals()` that returns the total number of animals created.
8. Add a `@staticmethod` named `is_valid_name(name)` that returns `True` only if `name` is a string and, after stripping spaces, is not empty.
9. Add an abstract method named `speak()`.

Remember:

```python
from abc import ABC, abstractmethod
```

````python
# Part 2 Code Here


from abc import ABC, abstractmethod

# TODO: Build the Animal abstract class here
class Animal(ABC):
  total_animals = 0
  def __init__(self, name, age):
    self.name = name
    self.age = age
    Animal.total_animals += 1

  def show_info(self) -> str:
    return f"{self.name} is {self.age} years old."

  @classmethod
  def get_total_animals(cls):
    return f"Total number of animals: {cls.total_animals}"

  @staticmethod
  def is_valid_name(name):
    if isinstance(name, str):
      if name.strip() != "":
        return True
      return False
    else:
      return False


  @abstractmethod
  def speak(self):
    return "I speak."
````

# Part 7 — Abstract Contract Debugging

The following code violates an abstract method contract.

Run it, observe the error, then answer the questions and fix it in the next cell.

````python
# Part 7 Code — This code intentionally violates the abstract contract

from abc import ABC, abstractmethod

class Vehicle(ABC):
    def __init__(self, brand):
        self.brand = brand

    def show_brand(self):
        return f"Brand: {self.brand}"

    @abstractmethod
    def move(self):
        pass

class Car(Vehicle):
    pass

class Bicycle(Vehicle):
    def move(self):
        return "Pedaling forward"

# This should produce an error because Car does not implement move().
my_vehicles = [Car("Toyota"), Bicycle("Trek")]
````

### Part 7 Questions

1. What error did Python produce?
2. Why did Python produce that error?
3. Why is `show_brand()` allowed to be inherited as-is?
4. Why must `move()` be implemented in the child class?

**Answers:**

1. Car cannot instantiate abstract class Car without implemeting the abstract method move().

2. Every child is required to implement the abstract method. If it does not, it produces an error.

3. That is an instance method not abstract so it uses the attributes and self, referring to the specific object, allowing each child to use that method without calling again.

4. Move() is abstract and an abstract method must be implemented in every child class.

````python
# Part 7 Fixed Code Here

# TODO: Fix Car so that it follows the abstract contract.
# Then create a list of Vehicle objects and loop through them polymorphically.

from abc import ABC, abstractmethod

class Vehicle(ABC):
    def __init__(self, brand):
        self.brand = brand

    def show_brand(self):
        return f"Brand: {self.brand}"

    @abstractmethod
    def move(self):
        pass

class Car(Vehicle):
    def move(self):
        return "Vrooommm"

class Bicycle(Vehicle):
    def move(self):
        return "Pedaling forward"

my_vehicles = [Car("Toyota"), Bicycle("Trek")]
for vehicle in my_vehicles:
  print(vehicle.show_brand())
  print(vehicle.move())
````

# Part 9 — Overriding vs Reassignment vs Shadowing

````python
class Parent:
    SOUND = "generic sound"

    def __init__(self):
        self.mood = "calm"

    def speak(self):
        return "Parent speaking"

class Child(Parent):
    SOUND = "child sound"

    def __init__(self):
        super().__init__()
        self.mood = "excited"

    def speak(self):
        return "Child speaking"

p = Parent()
c = Child()

print(Parent.SOUND)
print(Child.SOUND)
print(p.mood)
print(c.mood)
print(p.speak())
print(c.speak())
````

## Part 12 — Multiple Inheritance with a Mixin

In Python, a class can inherit from more than one parent class. This is called **multiple inheritance**.

A common safe use of multiple inheritance is a **mixin**.

A mixin is a small helper class that adds one focused behavior to another class.

In this problem, you will create a mixin called `AdoptableMixin` and use it with one of your `Animal` child classes.

### Important Idea

- `Animal` provides the main identity and required abstract behavior.
- `AdoptableMixin` provides one extra reusable behavior.
- The final child class can inherit from both.

Example structure:

```python
class AdoptableDog(Dog, AdoptableMixin):
    ...
```

### Hint

````python
class AdoptableMixin:
  def __init__(self):
    self.adopted = False

  def adopt(self):
    self.adopted = True

  def is_adopted(self):
    return self.adopted

class AdoptableDog(Dog, AdoptableMixin):
  def __init__(self, name, age, breed):
    Dog.__init__(self,name, age, breed)
    AdoptableMixin.__init__(self)

AD1 = AdoptableDog("Pal", 3, "Poodle")
AD1.adopt()
print(AD1.is_adopted())
print(AD1.show_info())
print(AD1.speak())
````
