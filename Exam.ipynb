```python
def process_grades(input_file, output_file):
  student_grade = {}
  with open(input_file, "r") as input:
    lines = input.readline()
    while lines != "":
      strip_line = lines.strip()
      split_line = strip_line.split(",")
      name = split_line[0]
      subject = split_line[1]
      grade = split_line[2]

      try:
        grade = int(grade)
        if name in student_grade:
          student_grade[name][subject] = grade
        else:
          student_grade[name] = {subject: grade}
        lines = input.readline()

      except ValueError:
        lines = input.readline()


  with open(output_file, "w") as output:
      total = 0
      count = 0
      greatest = student_grade[name][subject]
      greatest_name = name
      for name, v in student_grade.items():
        for subject, grade in v.items():
          total += int(grade)
          count +=1
        average = total / count
        if average > greatest:
          greatest = average
          greatest_name = name

        output.write(f"{name}: {average}\n")
        total = 0
        count = 0

  return greatest_name
```

```python
def sum_valid_numbers(data):
  if len(data) == 0:
    return 0

  if isinstance(data[0], int) and not isinstance(data[0], bool):
    count = data[0] + sum_valid_numbers(data[1:])
  else:
    count = sum_valid_numbers(data[1:])

  return count

data = [5, "hello", 3, True, 7, "9", 2]
sum_valid_numbers(data)
```

```python
from abc import ABC, abstractmethod
class GameCharacter(ABC):
  num_characters = 0
  def __init__(self, name, level):
    self._name = name
    self._level = level
    GameCharacter.num_characters += 1

  @staticmethod
  def check_experience(level):
    if level > 10:
      return True
    else:
      return False

  @abstractmethod
  def show_player(self):
    pass

  #method overriding
  def reaction(self):
    return "YAY!"

  @classmethod
  def get_num(cls):
    return cls.num_characters

class Hero(GameCharacter):
  def __init__(self, name, level, age):
    super().__init__(name, level)
    self._age = age

  def show_player(self):
    return f"{self._name} is at level {self._level}"

  def reaction(self):
    return "Oh my!"

class Villian(GameCharacter):
  def __init__(self, name, level, strength):
    super().__init__(name,level)
    self._strength = strength

  def show_player(self):
    return f"{self._name} is at level {self._level}"

  def reaction(self):
    return "HAHAHAHA..."


class Sidekick(GameCharacter):
  def __init__(self, name, level, tall):
    super().__init__(name,level)
    self._tall = tall

  def show_player(self):
    return f"{self._name} is at level {self._level}"

  def reaction(self):
    return "No way!"



characters = [Hero("John", 85, 54), Villian("Greg", 72, 112), Sidekick("Bob", 45, 6)]
for char in characters:
  print(char.show_player())
  print(char.reaction())

#print(characters[0].check_experience(12))
#print(characters[2].get_num())
```

```python
class Course:
  def __init__(self, course_name, instructor, max_seats, enrolled_students):
    self.course_name = course_name
    self.instructor = instructor
    self.max_seats = max_seats
    self.enrolled_students = enrolled_students

  def add_student(self, name):
    if name not in self.enrolled_students and len(self.enrolled_students) != self.max_seats:
      self.enrolled_students.append(name)
      return True
    else:
      return False

  def remove_student(self, name):
    if name in self.enrolled_students:
      self.enrolled_students.remove(name)

  def get_enrollment_count(self):
    return len(self.enrolled_students)

  def display_course_info(self):
    return f"The {self.course_name} course is taught by {self.instructor} and has a max total of {self.max_seats} seats. The students enrolled in this course: {self.enrolled_students}"

def find_course_with_most_available_seats(courses):
  high = courses[0].max_seats - len(courses[0].enrolled_students)
  for course in courses:
    count = course.max_seats - len(course.enrolled_students)
    if count > high:
      high = count
      course_high = course

  return course_high

courses = [Course("CS", "Prof. Cohen", 12, ["Leah", "Ruchama"]), Course("History", "Prof. Katz", 20, ["Shaked", "Chani"]), Course("Math", "Prof. Stein", 20, ["Reizi"])]
find_course_with_most_available_seats(courses)
```

```python
def two_sum(nums, target):
  for a in range(len(nums)):
    for b in range(1, len(nums)):
      if nums[a]+ nums[b] == target:
        return (a,b)
  return None

two_sum([2, 7, 11, 15], 9)
```
