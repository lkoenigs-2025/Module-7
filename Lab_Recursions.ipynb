**Recursion**: breaking a problem up into smaller pieces

**Stack**: linear(like straight line) data stucture that follows rule LIFO (Can only get whats on top)

**Push**: take another piece of data and put on top of stack

**Pop**: remove top/ last item

**Peek**: view the top item

**CALL STACK**:

when functions are called every call to function gets pushed onto the call stack
stack frame: when each function is called it creates a stack frame and contians everything it needs to be able to call function: local variables, parameters, return location and excersion state...

**Recusrion process uses this stack frame**- pushes each function onto this call stack and then each call will store all the four above and we can make use of this!

**Base Case:** where it stops

*Winding process*: when function calls are repeatedly pushed onto the call stack

*Unwinding process*: begins when base case ends: as stack frames popped off

--2 kinds of unwinding--

1. Non productive: all computation occurs in the winding phase
2. Productive: the computation occurs in the unwinding

**Depth of recursion** = number of calls u have and when then the recursion ends (base case) the depth == 0

**RecursionError**: when excedes

```python
def message(count: int = 5):
  if count > 0:
    #WIND: goes to each function call
    print("This is our count: ", count)
    count-=1
    message(count)
    #base case: when reaches zero (in this case)
    #UNWIND: only when it hits 0 does this unwind happen
    print("This is current unwind:", count)

if __name__ == "__main__":
  message()
```

```python
def speak(count):
  #base case
  if count == 0:
    return

  print("WIND:", count)
  speak(count-1)
  print("UNWIND:", count)
speak(5)
#when unwind 1 is on top cuz its the lasst we put  on the stack
```

```python
def factorial(n):
  #1. figure out base case:
  if n == 0:
    return 1

  #i want to return the previous factorial num
  #it will keep calling factorial until gets to base case
  value = factorial(n-1) #here in factorial function we have n and n-1

  # will print 1 cuz base case returns 1, then 1 *1 = 1, then 2 * 1 = 2, then 3 * 2 = 6, then 4 *6 = 24, 5 * 24 = 120
  print("value:", value)
  #never have the entire stack
  return n * value

factorial(5)
```

```python
#regular looping
def sum_iterable(numlst: list):
  total = 0
  for value in numlst:
    total += value
  return total

numlst = [10,20,30,40,50]
sum_iterable(numlst)
```

```python
def sum_recur(numlst: list, index):

  #base case
  if index == 0:
    return numlst[0]

  value = sum_recur(numlst, index - 1)
  return numlst[index] + value #value of previuos and current


numlst = [10,20,30,40,50]
sum_recur(numlst, len(numlst) -1)
```

```python
def product(lst, index):
  #base case:
  if index == 0:
    return lst[0]

  value = product(lst, index -1)
  return lst[index] * value
lst = [2,3,4]
product(lst, len(lst)-1)
```

```python
def product(lst, index):
  #base case:
  if index == 0:
    if lst[0] % 2 == 0:
      return 1 #even
    else:
      return 0 #odd

  value = product(lst, index -1)
  if lst[index] % 2 == 0:
    return 1 + value
  else:
    return value

lst = [2, 3, 4, 5, 6, 7, 8]
product(lst, len(lst)-1)
```
