```python
class GuessValidator:
   """A small utility class for validating guesses."""

   @staticmethod
   def is_single_letter(text: str) -> bool:
       return len(text) == 1 and text.isalpha()

   @staticmethod
   def normalize_guess(text: str) -> str:
       return text.lower().strip()


raw_guess = " A "
guess = GuessValidator.normalize_guess(raw_guess)

print(guess)
print(GuessValidator.is_single_letter(guess))
print(GuessValidator.is_single_letter("ab"))
```

```python
from abc import ABC, abstractmethod

class WordGameController(ABC):
  @abstractmethod
  def play(self):
    pass

  @abstractmethod
  def make_guess(self):
    pass

  @abstractmethod
  def check_game_over(self):
    pass

class TinyWordGame(WordGameController):
  def __init__(self, secret_word):
    self.secret_word = secret_word
    self.guesses = set()

  def play(self):
    print("Welcome to the Tiny Word Game, Guess the Secret Letter")

  def make_guess(self, letter):
    self.guesses.add(letter)
    return letter in self.secret_word

  def check_game_over(self):
    #go thru every letter compare and see if every letter in self.guesses and if all in there returns true
    return all(letter in self.guesses for letter in self.secret_word)

game = TinyWordGame("flower")
game.play()
game.make_guess("c")
print(game.check_game_over())
```

```python
from abc import ABC, abstractmethod

class WordGameController(ABC):
  @abstractmethod
  def play(self):
    pass

  @abstractmethod
  def make_guess(self):
    pass

  @abstractmethod
  def check_game_over(self):
    pass

class HintWordGame(WordGameController):
  def __init__(self, secret_word, hint):
    self.secret_word = secret_word
    self.hint = hint
    self.guesses = set()

  def play(self):
    print("Hint:", self.hint, "in the word")

  def make_guess(self, letter):
    self.guesses.add(letter)
    return letter in self.secret_word

  def check_game_over(self):
    #go thru every letter compare and see if every letter in self.guesses and if all in there returns true
    return all(letter in self.guesses for letter in self.secret_word)

game = HintWordGame("flower", "f")
game.play()
game.make_guess("c")
print(game.check_game_over())
```

```python
class LimitedGuessWordGame(WordGameController):
  def __init__(self, secret_word, num_guesses):
    self.secret_word = secret_word
    self.num_guesses = num_guesses
    self.guesses = set()
    self.guess_count = 0

  def play(self):
    print(f"You have {self.num_guesses} guesses.")


  def make_guess(self, letter):
    self.guesses.add(letter)
    self.guess_count +=1
    return letter in self.secret_word

  def check_game_over(self):
    if self.guess_count >= self.num_guesses:
      return all(letter in self.guesses for letter in self.secret_word)

    else:
      return f"You have {self.num_guesses - self.guess_count} guesses left"

game = LimitedGuessWordGame("hello", 6)
game.play()
print(game.make_guess("h"))
print(game.check_game_over())

```

```python
class Flyer:
  def __init__(self, animal_name):
    self.animal_name  = animal_name

  @staticmethod
  def fly():
    print("I am flying")

class Swimmer:
  def __init__(self, swimmer_name):
    self.swimmer_name = swimmer_name

  @staticmethod
  def swim():
    print("I am swimming")

#can pass two classes
class FlySwimmer(Flyer, Swimmer):
  def __init__(self, animal_name, swimmer_name):
    #cant use super cuz u called two class and python will get confused
    #u need self as well
    Flyer.__init__(self, animal_name)
    Swimmer.__init__(self, swimmer_name)

duck = FlySwimmer("daffy", "swimming")
duck.swim()
duck.fly()
```

```python
from abc import ABC
class Vehicle(ABC):
  def __init__(self, brand, speed):
    self.brand = brand
    self.speed = speed

class Fly:
  def __init__(self, altitude, airport_destination):
    self.altitude = altitude
    self.airport_destination = airport_destination

class FlyVehicle(Fly, Vehicle):
  def __init__(self, brand, speed, altitude, airport_destination):
    Vehicle.__init__(self, brand, speed)
    Fly.__init__(self, altitude, airport_destination)


car = FlyVehicle("BMW", 95, 1999, "JFK")
car.speed

```
