```python
class Person:
  def __init__(self, name):
    self.name = name

  def speak(self, msg):
    print(msg)

  def walking(self):
    print("I am walking")

#derived class
class Student(Person):
  def __init__(self, name, age):
    #store a new name for parent class but wont be used if dont call parent class
    super().__init__("Ruchama")
    self.name = name
    self.age = age

  def study(self):
    print("I am studying")

  #override: the interpreter will start with child then go up to parent so parent overriden
  def speak(self, msg):
    print("In student class:", msg)

    #access the parent version by saying super()
    super().speak("Good morning, parent speaking")


s1 = Student("Leah", 19)

s1.speak("Good Morning!")
print()
s1.study()

#access to parent class person
s1.walking()

#attributesof student
print(s1.name)
print(s1.age)
```
