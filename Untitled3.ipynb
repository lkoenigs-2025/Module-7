```python
def find_max(nums):
  high = nums[0]
  for i in nums:
    if i > high:
      high = i
  return high

print(find_max([1,3,53,6,4,3,46,2]))
```

```python
def average(nums):
  total = 0
  for i in nums:
    total += i

  average = total / len(nums)
  return average

print(average([1,3,53,6,4,3,46,2]))
```

```python
def split_even_odd(nums):
  evens = []
  odds = []
  for i in nums:
    if i % 2 == 0:
      evens.append(i)
    else:
      odds.append(i)
  return (evens, odds)

print(split_even_odd([1,3,53,6,4,3,46,2]))
```

```python
def largest_in_matrix(matrix):
  largest = 0
  for i in matrix:
    for j in i:
      if j > largest:
        largest = j
  return largest
```

```python
def student_report(grades):
  count = 0
  for key, value in grades.items():
    for v in value:
      count += v
    average = count / len(value)
    grades[key] = f"{average:.0f}"
    count = 0
  return grades

student_report({
    "Leah":[90,95,100],
    "Sarah":[80,70,90]
})

# {
#   "Leah":95,
#   "Sarah":80
# }
```

```python
from string import punctuation
def most_common_word(sentence):
  new_dict = {}
  greater = 0
  new_string = ""
  punctuation = "!^():?><,./;[]{}"

  for i in sentence:
    if i not in punctuation:
      new = i.lower()
      new_string += new

  words = new_string.split()

  for word in words:
    if word not in new_dict:
      new_dict[word] = 1

    else:
      new_dict[word] += 1
  for key in new_dict.keys():
    if new_dict[key] > greater:
      greater = new_dict[key]
      greatest_word = key
  return greatest_word

most_common_word(
    "Dog. cat:: dog:... fish fish DOG"
)

# "dog"
```

```python
def matrix_stats(matrix):
  largest = matrix[0][0]
  smallest = matrix[0][0]
  total = 0

  for i in matrix:
    for j in i:
      total += j

      if j > largest:
        largest = j

      if j < smallest:
        smallest = j

  return (total, largest, smallest)

matrix_stats([
    [1,2,3],
    [4,5,6]
])

# (21,6,1)
```

```python
def first_unique_word(sentence):
  dictw = {}
  words = sentence.split()
  for w in words:
    if w not in dictw:
      dictw[w] = 1
    else:
      dictw[w] +=1

  for key in dictw.keys():
    if dictw[key] == 1:
      return key

  return None


print(first_unique_word(
    "cat dog cat fish dog"
))

# "fish"

print(first_unique_word(
    "a a b b"
)
)
# None
```

```python
def group_students(scores):
  scores_new = {}
  A = 90
  B = 80
  C = 70

  for key, value in scores.items():
    if value >= A:
      if "A" not in scores_new:
        scores_new["A"] = [key]
      else:
        scores_new["A"].append(key)
    elif value >= B:
      if "B" not in scores_new:
        scores_new["B"] = [key]
      else:
        scores_new["B"].append(key)
    elif value >= C:
      if "C" not in scores_new:
        scores_new["C"] = [key]
      else:
        scores_new["C"].append(key)
    else:
      if "D" not in scores_new:
        scores_new["D"] = [key]
      else:
        scores_new["D"].append(key)

  return scores_new


group_students({
    "Leah":95,
    "Sarah":82,
    "Rachel":61
})
```

```python
def mutual_friends(network, person1, person2):
  for friend in network.keys():
    if friend == person1:
      person1 = set(network[friend])
    elif friend == person2:
      person2 = set(network[friend])
  return person1.intersection(person2)


network = {
    "Leah":{"Sarah","Rachel","Miriam"},
    "Sarah":{"Leah","Rachel"}
}

mutual_friends(
    network,
    "Leah",
    "Sarah"
)

# {"Rachel"}
```

```python
from abc import ABC, abstractmethod
class Creature:
  creature_count = 0

  def __init__(self, name, energy):
    self.name = name
    self.energy = energy
    Creature.creature_count +=1

  @abstractmethod
  def perform_move(move: str):
    pass

  @abstractmethod
  def show_status():
    pass

  @classmethod
  def get_creature_count(cls):
    return cls.creature_count


class Dragon(Creature):
  def __init__(self, name, energy,fire_power):
    super().__init__(name,energy)
    self.fire_power = fire_power

  @abstractmethod
  def perform_move(self, move: str):
    move = move.lower().strip()
    if move != "fireball" and move != "fly" and move !="roar":
      print("Invalid move.")
      return False
    else:
      if self.energy >=10:
        self.energy -= 10
        print(f"{self.name} uses {move}!")
        return True
      else:
        "Not enough energy."
        return False

  @abstractmethod
  def show_status(self):
    print(f"Name: {self.name} \nEnergy: {self.energy} \nFire Power: {self.fire_power}")

d = Dragon("Draco", 100, 75)

d.perform_move("fireball")

d.perform_move("jump")

d.show_status()
```
