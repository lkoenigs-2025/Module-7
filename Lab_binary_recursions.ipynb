```python
#O(n) we have to search each
def linear_search_itr(nlist, target):
  for index, value in enumerate(nlist):
    if value == target:
      print(f"Item found at index {index} value {value}")
      return True
  return False

print(linear_search_itr([20,30,40,50], 50))
print(linear_search_itr([20,30,40,50], 35))
```

```python
def linear_search(nlist, target, index):
  if index == 0:
    return nlist[index]


  if nlist[index-1] == target:
    print("Item found in list")
  linear_search(nlist, target, index-1)
  return False


linear_search([20,30,40,50], 20, 4)
linear_search([20,30,40,50], 35, 4)

```

```python
#binary search

def binary_search(nlist, target, low, high):
  #base case 1:
  if low > high:
    return -1 #didnt find

  #take midpoint
  mid = (low + high) // 2

  #base case 2:
  if nlist[mid] == target:
    return mid

  #recursive case 1:
  #if target less than midpoint
  elif target < nlist[mid]:
    return binary_search(nlist, target, low, mid-1)
  #recursive case 2
  #if target greater than mid
  else:
    return binary_search(nlist, target, mid+1, high)


nlist = [10,20,30,40,50,60]
print(binary_search(nlist, 30, 0, len(nlist)-1))
```

```python
#splitting: all recursion occurs
def merge_sort(nums):

  #base case
  if len(nums) <= 1:
    return nums

  mid = len(nums) //2

  left = merge_sort(nums[:mid])    #up to but not incl midpoint
  right = merge_sort(nums[mid:])   #including mid and then to end of all data

  return merge(left, right)


def merge(left,right):
  merged = []

  left_index = 0
  right_index = 0

  while left_index < len(left) and right_index < len(right):
    if left[left_index] < right[right_index]:
      merged.append(left[left_index])
      left_index +=1
    else:
      merged.append(right[right_index])
      right_index +=1

  while left_index < len(left):
    merged.append(left[left_index])
    left_index +=1

  while right_index < len(right):
    merged.append(right[right_index])
    right +=1

  return merged


nlist = [50,40,30,20,10]
print(merge_sort(nlist))
```
