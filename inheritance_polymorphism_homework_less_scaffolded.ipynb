## Problem 1: Build an Animal Hierarchy

Create a parent class named `Animal`.

### The `Animal` class must:
- store a name
- store an age
- include a method named `speak()` that returns a general animal sound
- include a method named `move()` that returns a general description of movement
- include a method named `describe()` that returns a short sentence about the animal

Then create these child classes:

- `Dog`
- `Cat`
- `Rabbit`

### Child class requirements
Each child class must:
- inherit from `Animal`
- use the parent constructor appropriately
- override `speak()`
- override `move()`

### After creating your classes
Create one object of each child class and demonstrate that:
- each object stores its own data
- each object gives a different result for `speak()`
- each object gives a different result for `move()`
- `describe()` works for each object


```python
# Write your code for Problem 1 here.
class Animal:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def speak(self):
    return "Raaaaa"

  def move(self):
    return f"Skipping"

  def describe(self):
    return f"I love to sleep"

class Dog(Animal):
  def __init__(self):
    super().__init__("Dog", 3)

  def speak(self):
    return "Woof!"

  def move(self):
    return f"I am running!!"


class Cat(Animal):
  def __init__(self):
    super().__init__("Cat", 2)

  def speak(self):
    return "Meow."

  def move(self):
    return f"I am sleeping."

class Rabbit(Animal):
  def __init__(self):
    super().__init__("Rabbit", 1)

  def speak(self):
    return "I'm quiet..."

  def move(self):
    return f"I am climbing."
```

```python
#create 3 objects
dog = Dog()
cat = Cat()
rabbit = Rabbit()

print(f"My {dog.name} is {dog.age} years old.")
print(f"My {cat.name} is {cat.age} years old.")
print(f"My {rabbit.name} is {rabbit.age} years old.\n")

#speak methos overriden in each child class
print(dog.speak())
print(cat.speak())
print(rabbit.speak(),"\n")

#move method overriden in each child class
print(dog.move())
print(cat.move())
print(rabbit.move(),"\n")

#describe method works for each method
print(dog.describe())
print(cat.describe())
print(rabbit.describe())
```

## Problem 2: Build a Vehicle Hierarchy

Create a parent class named `Vehicle`.

### The `Vehicle` class must:
- store a brand
- store a speed
- include a method named `move()` that returns a general statement about transportation
- include a method named `sound_horn()` that returns a general horn sound
- include a method named `describe()` that returns a short sentence about the vehicle

Then create these child classes:

- `Car`
- `Truck`
- `Motorcycle`

### Child class requirements
Each child class must:
- inherit from `Vehicle`
- use the parent constructor appropriately
- override `move()`
- override `sound_horn()`

### After creating your classes
Create one object of each child class and demonstrate that:
- each object stores its own data
- each object gives a different result for `move()`
- each object gives a different result for `sound_horn()`
- `describe()` works for each object


```python
# Write your code for Problem 2 here.
class Vehicle:
  def __init__(self, brand, speed):
    self.brand = brand
    self.speed = speed

  def move(self):
    return "Get ready, get set, gooooo!"

  def sound_horn(self):
    return "Beeeeeep"

  def describe(self):
    return "Pay for the trip."

class Car(Vehicle):
  def __init__(self):
    super().__init__("Toyota", 55)

  def move(self):
    return "Fasten your seatbelts!"

  def sound_horn(self):
    return "Beep beep"

class Truck(Vehicle):
  def __init__(self):
    super().__init__("Ford", 32)

  def move(self):
    return "Turn on your headlights!"

  def sound_horn(self):
    return "HONNNNKK!!"

class Motorcycle(Vehicle):
  def __init__(self):
    super().__init__("Honda", 64)

  def move(self):
    return "Keep out of the way!!"

  def sound_horn(self):
    return "EEEEE"
```

```python
#create 3 objects
car = Car()
truck = Truck()
motorcycle = Motorcycle()

print(f"My {car.brand} is going {car.speed} mph.")
print(f"My {truck.brand} is going {truck.speed} mph.")
print(f"My {motorcycle.brand} is going {motorcycle.speed} mph.\n")

#speak methos overriden in each child class
print(car.sound_horn())
print(truck.sound_horn())
print(motorcycle.sound_horn(),"\n")

#move method overriden in each child class
print(car.move())
print(truck.move())
print(motorcycle.move(),"\n")

#describe method works for each method
print(car.describe())
print(truck.describe())
print(motorcycle.describe())
```


## Problem 3: Demonstrate Polymorphism with Animals

Using your `Dog`, `Cat`, and `Rabbit` classes:

1. Create a list that stores multiple animal objects together.
2. Use a loop to go through the list.
3. For each object, call the **same methods** and show that the results differ depending on the object.

### Required methods to call in the loop
- `speak()`
- `move()`
- `describe()`

## Goal
Your output should clearly demonstrate that the program can treat all of the child objects as `Animal` objects, while each one still behaves in its own way.


```python
# Write your code for Problem 3 here.
list_of_animals = [Dog(), Cat(), Rabbit()]
for animal in list_of_animals:
  print(animal.speak())
  print(animal.move())
  print(animal.describe(), "\n")
```
