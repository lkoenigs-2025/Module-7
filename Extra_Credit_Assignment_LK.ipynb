```python
from abc import ABC, abstractmethod

class Student(ABC):
  total_num_students = 0
  def __init__(self, name, grade):
    self._name = name
    self._grade = grade
    Student.total_num_students +=1

  @abstractmethod
  def reaction_to_test(self):
    pass

  @classmethod
  def get_total_students(cls):
    return cls.total_num_students

  @staticmethod
  def check_grade(grade):
    if grade < 60:
      return "Error! Grade is too low."
    else:
      return "Great!"

class Math(Student):
  def __init__(self, name, grade):
    super().__init__(name, grade)

  def reaction_to_test(self):
    if self._grade > 85:
      return "Yipeeee!!!"
    else:
      return "Okay..."


class Science(Student):
  def __init__(self, name, grade):
    super().__init__(name, grade)

  def reaction_to_test(self):
    return "Woah!"

class History(Student):
  def __init__(self, name, grade):
    super().__init__(name, grade)

  def reaction_to_test(self):
    return "Oh my!!"




m1 = Math("Leah", 92)
s1 = Science("Sarah", 88)
h1 = History("Rachel", 100)

students = [m1, s1, h1]
for student in students:
  print(student.reaction_to_test())
```
