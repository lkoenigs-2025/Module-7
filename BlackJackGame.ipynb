```python
#Represents one playing card
class Card:
    def __init__(self, suit: str, rank: str, value: int):
        self.suit = suit
        self.rank = rank
        self.value = value

    def __str__(self)-> str:
        return f"{self.rank} of {self.suit}"
```

```python
import random
#represents the full deck of cards
class Deck:
    def __init__(self):
        self.cards = []

    #build the deck of cards
    def build_deck(self) -> None:
        suits = ["Hearts", "Diamonds", "Clubs", "Spades"]
        values = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
        ranks = ["Ace", "2", "3", "4", "5", "6", "7", "8", "9", "10", "Jack", "Queen", "King"]

        for suit in suits:
            for card in range(len(values)):
                self.cards.append(Card(suit, ranks[card], values[card]))

    #shuffle deck
    def shuffle(self) -> None:
        random.shuffle(self.cards)

    #deal a card in deck
    def deal_card(self) -> Card:
        return self.cards.pop()

    #find numbers of cards remaining
    def cards_remaining(self) -> int:
        return len(self.cards)
```

```python
#represents the cards held by one participant
class Hand:
    def __init__(self):
        self.cards = []

    #add card
    def add_card(self, card: Card)-> None:
        self.cards.append(card)

    #total of cards in hand
    def get_total(self) -> int:
        total = 0
        aces = 0

        for card in self.cards:
            total += card.value

            if card.rank == "Ace":
                aces +=1

        while total > 21 and aces > 0:
            total -= 10
            aces -= 1

        return total

    #show cards
    def show_hand(self) -> str:
        string = ""
        for card in self.cards:
            string += f"{card.rank} of {card.suit} | "

        return string

    def __str__(self)-> str:
        return f"Hand: {self.show_hand()} (Total: {self.get_total()})"
```

```python
#Parent class for Player and Dealer
class Participant:
    def __init__(self, name: str):
        self.name = name
        self.hand = Hand()

    def take_card(self, card: Card) -> None:
        self.hand.add_card(card)

    #show what cards have
    def show_hand(self) -> str:
        return self.hand.show_hand()

    #determine total of cards
    def get_total(self) -> int:
        return self.hand.get_total()

    #determine if player hand is greater than 21
    def is_busted(self) -> bool:
        if self.hand.get_total() > 21:
            return True
        else:
            return False

    #will be overridden by player and dealer class
    def take_turn(self, deck: Deck) -> None:
        self.take_card(deck.deal_card())
```

```python
#child class of participant
class Player(Participant):
    def __init__(self, name: str):
        super().__init__(name)

    #overrides take_turn()
    def take_turn(self, deck: Deck) -> None:
        while not self.is_busted():
            play = input("\nHit (h) or Stand (s)?" )

            if play.lower() == "h":
              self.take_card(deck.deal_card())
              print(f"\n{self.show_hand()}")
              print(f"\nTotal: {self.get_total()}")

            elif play.lower() == "s":
                break

            else:
                print("\nHit (h) or Stand (s)?" )
```

```python
#child class of participant
class Dealer(Participant):
    def __init__(self):
        super().__init__("Dealer")

    #show the first card of dealer
    def show_first_card(self) -> str:
        return str(self.hand.cards[0])

    #dealer takes a turn
    def take_turn(self, deck: Deck) -> None:
        while self.get_total() < 17:
            self.take_card(deck.deal_card())
```

```python
#controller class for the game
class BlackJackGame:
    def __init__(self, player_name: str):
        self.deck = Deck()
        self.player = Player(player_name)
        self.dealer = Dealer()

    #start game
    def initial_deal(self) -> None:
        self.deck.build_deck()
        self.deck.shuffle()
        for participant in [self.player, self.dealer]:
            participant.take_card(self.deck.deal_card())
            participant.take_card(self.deck.deal_card())

    #show what game is up to
    def show_game_state(self) -> None:
        print(f"\n{self.player.name}'s hand: {self.player.show_hand()}"
              f"\nTotal: {self.player.get_total()}")
        print(f"\nDealer's first card: {self.dealer.show_first_card()}")

    #finds the winner
    def determine_winner(self) -> str:
        player_total = self.player.get_total()
        dealer_total = self.dealer.get_total()


        if player_total > 21:
            return "\nDealer wins!"
        elif dealer_total > 21:
            return "\nPlayer wins!"
        elif player_total > dealer_total:
            return "\nPlayer wins!"
        elif player_total < dealer_total:
            return "\nDealer wins!"
        else:
            return "\nTie"

    #function that plays the game
    def play(self) -> None:
        self.initial_deal()
        self.show_game_state()

        print("\n------------------------------\n")
        self.player.take_turn(self.deck)

        if not self.player.is_busted():
            print("\n------------------------------\n")
            self.dealer.take_turn(self.deck)
            print(f"\nDealer's hand: {self.dealer.show_hand()}")
            print(f"\nDealer's total: {self.dealer.get_total()}")

        print("\n------------------------------\n")
        print(f"{self.player.name}'s hand: {self.player.show_hand()}"
              f"\nTotal: {self.player.get_total()}")
        print(f"\nDealer's hand: {self.dealer.show_hand()}"
              f"\nTotal: {self.dealer.get_total()}")
        print(self.determine_winner())
```
